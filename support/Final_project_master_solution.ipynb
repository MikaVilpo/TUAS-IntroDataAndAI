```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# pip install ucimlrepo
```

```python
# Load the dataset
#data = pd.read_csv("bank_marketing_dataset.csv")


from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
bank_marketing = fetch_ucirepo(id=222) 
data = bank_marketing 
# data (as pandas dataframes) 
X = bank_marketing.data.features 
y = bank_marketing.data.targets 
  
y1 = np.array(y).reshape(-1)
print(y1.shape) 

# variable information 
print(bank_marketing.variables) 
feature_names = data.feature_names

```

```python
# check number of missing values

missing_values = X.isnull().sum()
print(missing_values)

X.info()
```

```python
# Example: Impute missing values in a categorical column with the mode

X['education'].fillna(X['education'].mode()[0], inplace=True)
X['job'].fillna(X['job'].mode()[0], inplace=True)
X['contact'].fillna(X['contact'].mode()[0], inplace=True)
X = X.drop("poutcome", axis=1)
```

```python
# Explore the data
print(X.head())  # Display the first few rows
print(X.info())  # Get data type information
print(X.describe())  # Get summary statistics
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y1, test_size=0.2, random_state=42)
```

```python
# Encoding categorical features (one-hot encoding)


X_train = pd.get_dummies(X_train, columns=["job", "marital", "education", "default", "housing", "loan", "contact", "month", "day_of_week"])
X_test = pd.get_dummies(X_test, columns=["job", "marital", "education", "default", "housing", "loan", "contact", "month", "day_of_week"])
```

```python
# Data Scaling (normalize numerical features)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train.hist(bins=50, figsize=(100, 80))
plt.show()
```

```python
# Define the indices of the feature pairs you want to plot
feature_indices_to_plot = [(0, 1), (1, 2), (2, 3)]  # Example feature pairs

# Verify data structure and indices
if isinstance(X_train_scaled, (list, tuple, np.ndarray)):
    for feature1_idx, feature2_idx in feature_indices_to_plot:
        if 0 <= feature1_idx < X_train_scaled.shape[1] and 0 <= feature2_idx < X_train_scaled.shape[1]:
            # Create scatter plots for the selected feature pairs
            fig, axes = plt.subplots(1, len(feature_indices_to_plot), figsize=(15, 5))
            
            for i, (feature1_idx, feature2_idx) in enumerate(feature_indices_to_plot):
                axes[i].scatter(X_train_scaled[:, feature1_idx], X_train_scaled[:, feature2_idx])
                axes[i].set_xlabel(f'Feature {feature1_idx}')
                axes[i].set_ylabel(f'Feature {feature2_idx}')
                axes[i].set_title(f'Scatter Plot: Features {feature1_idx} and {feature2_idx}')
            
            # Show the plot
            plt.tight_layout()
            plt.show()
        else:
            print(f"Invalid feature indices: {feature1_idx} or {feature2_idx}")
else:
    print("X_train_clean is not a valid data structure (NumPy array or DataFrame).")
```

```python
# Replace 'df' with your actual DataFrame variable
df = pd.DataFrame(X_train_scaled)
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap for the scaled data
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

### **Step 2: Training Three Classification Models** (1.5 grade units)


**Model Selection:** Choose three classification algorithms: K-Nearest Neighbors (KNN), Logistic Regression, and Decision Tree. (0.5 grade units)

**Model Training:** ** Train each of the selected models on the training dataset. (0.5 grade units)

**Model Evaluation:** Evaluate the performance of each model on the testing dataset using appropriate evaluation metrics (e.g., accuracy, precision, recall, F1-score, ROC-AUC). (0.5 grade units)

```python
# Step 2: Training Three Classification Models
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
# Initialize and train the models
knn_model = KNeighborsClassifier()


knn_model.fit(X_train_scaled, y_train)
y_pred_knn = knn_model.predict(X_test_scaled)
```

```python
# Calculate accuracy as a measure of model performance
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred_knn)

print(f"Knn Model Accuracy: {accuracy:.2%}")
```

```python

logistic_model = LogisticRegression()


logistic_model.fit(X_train_scaled, y_train)
y_pred_logistic = logistic_model.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred_logistic)

print(f"Logistic Regression Model Accuracy: {accuracy:.2%}")

```

```python
decision_tree_model = DecisionTreeClassifier()

decision_tree_model.fit(X_train, y_train)

y_pred_tree = decision_tree_model.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred_tree)

print(f"Decision Tree Model Accuracy: {accuracy:.2%}")
```
