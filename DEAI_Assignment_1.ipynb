Import needed modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from ucimlrepo import fetch_ucirepo
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
# Load the dataset

# fetch dataset
bank_marketing = fetch_ucirepo(id=222)

# data (as pandas dataframes)
X = bank_marketing.data.features
Y = bank_marketing.data.targets
```

```python
# Output X to play with Data Wranger

X
```

```python
# Output Y to play with Data Wranger

Y
```

```python
# Explore the data

# Display the first few rows
print(X.head())

# Get data type information
print("\nData types")
print(X.info())

# Get summary statistics
print("\nDescribe Data")
print(X.describe())

# BEGIN: Loop through all object type fields and print description
for column in X.select_dtypes(include=['object']).columns:
    print(f"\nDescribe: {column}")
    print(X[column].describe())
# END:

# Check for missing values
print("\nMissing Values")
print(X.isnull().sum())

# Check for unique values in each column
print("\nUnique Values")
print(X.nunique())

```

```python
# Print out the dataframe columns and their dtypes.
print("Name, dtype, unique values")
for column in X.columns:
    print(column, X[column].dtype, X[column].nunique())
```

```python
# Handle missing values (e.g., replace missing values with the mean of the column)
# Split the columns between categorical and numerical values.
numerical_columns_list = ["day_of_week", "pdays"] # Keep these values as they are
normalized_columns_list = ["age", "balance", "duration", ] # Normalize these
categorical_columns_list = ["job", "marital", "education", "default", "housing", "loan", "contact",  "month", "campaign", "previous", "poutcome"] # Categorical values need to be encoded
```

```python
# Check the unique values of each categorical column
for column in categorical_columns_list:
    print(X[column].unique())
```

```python
# Handle poutcome seperately
# Add a label for poutcome to indicate "no contact"
X["poutcome"] = X["poutcome"].fillna("no_contact")

# Loop through all object type fields and replace missing values with the mode (value that appears most often)
for column in categorical_columns_list:
    mode = X[column].mode()
    # NOTE! This hack must been so that we only get the value of the mode.
    # As mode() returns a Series, it cannot be used for replacing nan values as itself.
    print(mode[0])
    X[column] = X[column].fillna(mode[0])

# Loop through all number type fields and replace missing values with the mean
for column in numerical_columns_list:
    X[column] = X[column].fillna(X[column].mean())

# Do the same with the normalized columns:
for column in normalized_columns_list:
    X[column] = X[column].fillna(X[column].mean())
```

```python
# Check the missing values in each dataset now
print("Training set missing values\n", X.isnull().sum())
```

```python
# All good!
```

```python
# Split data into training and testing sets and enable typing
training_and_testing_data: list[pd.DataFrame] = train_test_split(X, Y, test_size=0.2, random_state=24)
X_train, X_test, Y_train, Y_test = training_and_testing_data

# Display the first few rows of the training set
print("\nTraining Set")
print(X_train.head())

# Count the number of rows in the training set
print("\nTraining Set Rows")
print(X_train.shape[0])

# Display the first few rows of the testing set
print("\nTesting Set")
print(X_test.head())

# Count the number of rows in the testing set
print("\nTesting Set Rows")
print(X_test.shape[0])
```

```python
# Reindex both the training and testing data.
X_train = X_train.reset_index(drop=True).sort_index()
X_test = X_test.reset_index(drop=True).sort_index()
Y_train = Y_train.reset_index(drop=True).sort_index()
Y_test = Y_test.reset_index(drop=True).sort_index()
```

```python
# Encoding categorical features (e.g., one-hot encoding)

# Once more get and print out the categorical columns
categorical_columns = X_train[categorical_columns_list]
categorical_columns
```

```python
# Same with testing data
categorical_columns_test = X_test[categorical_columns_list]
categorical_columns_test
```

```python
# Get the normalized columns for the future
normalized_columns = X_train[normalized_columns_list]
normalized_columns
```

```python
# Same with test data
normalized_columns_test = X_test[normalized_columns_list]
normalized_columns_test
```

```python
# Same with numerical columns
numerical_columns = X_train[numerical_columns_list]
numerical_columns
```

```python
# Same with test data
numerical_columns_test = X_test[numerical_columns_list]
numerical_columns_test
```

```python
# Check unique number of categorical values in each column
categorical_columns.nunique()
```

```python
# Recheck what kind of unique values we have in each
# (We did this previously, but just to remind us)
for column in categorical_columns:
    print()
    print(column, ": ", categorical_columns[column].unique())
```

```python
# Define a label encoder
# We will use this to label encode some of the categorical data that has large number of values in it.
# We will later use one-hot encoding for the ones with less variables
label_encoder = preprocessing.LabelEncoder()
```

```python
# We use fit_transform() here but we need to use just "transform()" later for the test data
# This is done in order to avoid a bias
for column in categorical_columns[["job", "month"]]:
    categorical_columns[column] = label_encoder.fit_transform(categorical_columns[column])
    categorical_columns_test[column] = label_encoder.transform(categorical_columns_test[column])

categorical_columns, categorical_columns_test
```

```python
categorical_columns.sort_index()
```

```python
categorical_columns_test.sort_index()
```

```python
# Now use one-hot encoding for the rest.
# Define the one-hot encoder
one_hot_encoder = preprocessing.OneHotEncoder(sparse_output=False)
```

```python
# Process the one-hot encodable columns
one_hot_encoded_columns = ["marital", "education", "default", "housing", "loan", "contact", "poutcome"]
one_hot_encoded = one_hot_encoder.fit_transform(categorical_columns[one_hot_encoded_columns])
one_hot_encoded_test = one_hot_encoder.transform(categorical_columns_test[one_hot_encoded_columns])
```

```python
# Create a dataframe out of the one-hot encoded values
one_hot_df = pd.DataFrame(one_hot_encoded, columns=one_hot_encoder.get_feature_names_out(one_hot_encoded_columns))
one_hot_df_test = pd.DataFrame(one_hot_encoded_test, columns=one_hot_encoder.get_feature_names_out(one_hot_encoded_columns))

one_hot_df, one_hot_df_test
```

```python
# We can also encode the values in Y
Y_train = Y_train.replace({ "no": 0.0, "yes": 1.0 })
Y_test = Y_test.replace({ "no": 0.0, "yes": 1.0 })
Y_train, Y_test
```

```python
# Combine the dataframe with the previous one
categorical_columns = categorical_columns.drop(one_hot_encoded_columns, axis=1)
categorical_columns = pd.concat([categorical_columns, one_hot_df], axis=1)

categorical_columns_test = categorical_columns_test.drop(one_hot_encoded_columns, axis=1)
categorical_columns_test = pd.concat([categorical_columns_test, one_hot_df_test], axis=1)

categorical_columns, categorical_columns_test
```

```python
# Check that no missing values were generated
categorical_columns.isnull().sum(), categorical_columns_test.isnull().sum()
```

```python
# Data Scaling (normalize numerical features)
# Initialize scaler normalizers
normalizer = preprocessing.MinMaxScaler()
```

```python
# Normalize some of the numerical columns with a scaler
normalized_values = normalizer.fit_transform(normalized_columns)
normalized_values_test = normalizer.transform(normalized_columns_test)

normalized_values, normalized_values_test
```

```python
# Apply the normalized values to the dataframes
normalized_columns = pd.DataFrame(normalized_values, columns=normalized_columns.columns)
normalized_columns_test = pd.DataFrame(normalized_values_test, columns=normalized_columns_test.columns)

normalized_columns, normalized_columns_test
```

```python
# See how normalized values stack up
normalized_columns.hist(bins=50, figsize=(20,15))
normalized_columns_test.hist(bins=50, figsize=(20,15))
```

```python
# Combine now all the handled columns into a single dataframe
# Training
X_train = pd.concat([numerical_columns, categorical_columns, normalized_columns], axis=1)
X_train
```

```python
# Now combine testing
X_test = pd.concat([numerical_columns_test, categorical_columns_test, normalized_columns_test], axis=1)
X_test
```
